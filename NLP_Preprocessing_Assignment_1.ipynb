#### Task 1: Conceptual Understanding
Q1: What is the difference between "Love" and "love" in NLP?
In NLP, "Love" and "love" are technically different string tokens because computers are case-sensitive by default. Without lowercasing, a model treats them as two completely different words, inflating vocabulary size and reducing generalization. After lowercasing normalization, both map to "love", allowing the model to recognize them as the same concept regardless of sentence position.

Q2: What happens if stopwords are not removed?
If stopwords are not removed:

High-frequency noise: Words like "the", "is", "and" dominate frequency counts, drowning out meaningful words.
Increased dimensionality: The feature space grows unnecessarily, slowing training and increasing memory usage.
Reduced model accuracy: ML models learn spurious patterns from irrelevant tokens.
Poor TF-IDF scores: Stopwords get artificially inflated term frequencies, skewing vectorization.

Q3: Two real-world scenarios where removing stopwords can be harmful:
Sentiment Analysis with Negation:
"I am not happy with this product" becomes "happy product" after stopword removal — completely reversing the actual negative sentiment. The word "not" is a stopword but is semantically critical.

Question Answering / Information Retrieval:
A query like "Who is the president of India?" — removing stopwords strips "who", "is", "the" which are essential for understanding the query type and intent in Q&A systems.

Q4: Difference between Stemming and Lemmatization:

Stemming is a fast, rule-based technique that chops off word endings using heuristic patterns to reduce a word to its root form. The output may not always be a real dictionary word — for example, "running" becomes "runn" and "better" stays "better". Common algorithms include the Porter Stemmer and Snowball Stemmer. Because of its speed, stemming is widely used in search engines and quick text indexing tasks where precision is less critical.

Lemmatization, on the other hand, uses vocabulary and morphological analysis to reduce a word to its proper base form, called a lemma. The output is always a valid dictionary word — for example, "running" becomes "run" and "better" becomes "good". It relies on resources like the WordNet Lemmatizer and is slower than stemming, but significantly more accurate. Lemmatization is preferred in tasks that require deeper language understanding, such as chatbots, sentiment analysis, and natural language understanding (NLU) systems.

#### Task 2:Build Advanced Preprocessing Function

Implement the following function:

def preprocess_text(text):
    pass

Required Features

Your function must include:

Remove numbers
Example: "I have 2 dogs" → "I have dogs"

Remove extra spaces
Example: "This is good" → "This is good"

Handle repeated characters
Example: "soooo goooood!!!" → "so good"

Convert text to lowercase
Example: "WOW!!! This is GREAT!!!" → "wow this is great"

Remove very short tokens

Remove words with length ≤ 2

Exception: keep meaningful words like "no", "not"

Remove URLs and email-like patterns

Example: "Visit http://example.com now" → "visit now"


```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return []

    text = text.lower()
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens
```

```python
print("Test 1:", preprocess_text("I have 2 dogs"))
print("Test 2:", preprocess_text("soooo goooood!!!"))
print("Test 3:", preprocess_text("Visit http://example.com now"))
print("Test 4:", preprocess_text("WOW!!! This is GREAT!!!"))
print("Test 5:", preprocess_text("I am not happy with this"))
```

#### Task 4: Token Analytics
For each processed sentence, compute:

Total number of tokens

Number of unique tokens

Average token length

Analysis Questions

Which sentence had the most noise?

Which sentence retained the most meaningful tokens after cleaning?


```python
def analyze_tokens(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)

for s in sentences:
    tokens = preprocess_text(s)
    total, unique, avg = analyze_tokens(tokens)

    print(f"{s}")
    print(f"Total: {total}, Unique: {unique}, Avg Length: {avg}")
    print("-"*40)
```

#### Task 5: Frequency Analysis
Combine all tokens and use:

from collections import Counter

Top 10 most frequent words

```python
from collections import Counter

all_tokens = []

for s in sentences:
    all_tokens.extend(preprocess_text(s))

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nLeast 5 frequent words:")
print(counter.most_common()[-5:])
```

#### Task 6: Build Full Pipeline
def full_pipeline(text_list):
    pass

Expected Output
{
    "tokens": [...],
    "clean_sentences": [...]
}


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(sentences)
print(result)
```
