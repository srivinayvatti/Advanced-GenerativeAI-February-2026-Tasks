#### Problem 1: Smart Light Controller
Problem Statement
Create a class to control a smart light that can be turned ON or OFF and display its status.
🌍 Real-Time Use
Smart home automation systems
📥 Sample Input
Light Name: Bedroom Light
Action: ON
📤 Expected Output
Bedroom Light is ON


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True
        print(f"{self.name} is ON")

    def turn_off(self):
        self.is_on = False
        print(f"{self.name} is OFF")

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")
```

```python
bedroom_light = SmartLight("Bedroom Light")
bedroom_light.turn_on()
bedroom_light.display_status()

bathroom_light = SmartLight("Bathroom Light")
bathroom_light.turn_off()
bathroom_light.display_status()
```

#### Problem 2: Employee ID Card System
 Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.
Real-Time Use
Corporate HR systems
📥 Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT
📤 Expected Output
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT


```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
```

```python
rahul_card = EmployeeIDCard("Rahul", "EMP102", "IT")
rahul_card.display_id_card()
```

#### Problem 3: Mobile Contact Record
Problem Statement
Create a class to store contact name and phone number and display contact information.
Real-Time Use
Mobile contact applications
📥 Sample Input
Contact Name: Anita
Phone Number: 9876543210
📤 Expected Output
Contact Saved
Name: Anita
Phone: 9876543210


```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        print("Contact Saved")

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")
```

```python
anita_contact = Contact("Vinay", "9876543210")
anita_contact.display_contact()
```

#### Problem 4: Product Price Tag Generator
Problem Statement
Create a class that stores product name and price and prints a formatted price tag.
 Real-Time Use
Retail billing systems
📥 Sample Input
Product Name: Headphones
Price: 2499
📤 Expected Output
Product: Headphones
Price: ₹2499


```python
class ProductPriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
```

```python
headphones_tag = ProductPriceTag("Headphones", 2499)
headphones_tag.display_price_tag()
```

#### Problem 5: Movie Rating Display System
Problem Statement
Create a class that stores movie name and rating and displays movie details.
Real-Time Use
Streaming platforms
📥 Sample Input
Movie Name: Inception
Rating: 4.8
📤 Expected Output
Movie: Inception
Rating: 4.8 / 5


```python
class MovieRating:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
```

```python
inception_movie = MovieRating("Inception", 4.8)
inception_movie.display_movie_details()
```
