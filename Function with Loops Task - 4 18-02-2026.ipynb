#### Problem 1: Mobile Recharge Validation System
 Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry


```python
def validate_recharge_request():
    valid_plans = {199, 299, 399, 599}
    min_recharge_amount = 50

    while True:
        try:
            recharge_amount_str = input("Enter recharge amount (₹): ")
            recharge_amount = int(recharge_amount_str)

            if recharge_amount < min_recharge_amount:
                print(f"Invalid amount. Recharge amount must be at least ₹{min_recharge_amount}.\n")
                continue

            plan_str = input("Enter plan (199, 299, 399, 599): ")
            plan = int(plan_str)

            if plan not in valid_plans:
                print("Invalid plan. Please choose from 199, 299, 399, 599.\n")
                continue
            
            print("Recharge request is valid!\n")
            return {"amount": recharge_amount, "plan": plan}

        except ValueError:
            print("Invalid input. Please enter a valid number for amount and plan.\n")
```

```python
valid_request = validate_recharge_request()
print(f"Processed valid request: {valid_request}")
```

#### Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary

```python
def check_inventory(products_stock):
    reorder_threshold = 15
    print("--- Inventory Status ---")
    for product, stock in products_stock.items():
        if stock < reorder_threshold:
            print(f"{product}: Reorder Alert (Current stock: {stock})")
        else:
            print(f"{product}: Stock OK (Current stock: {stock})")
    print("------------------------")
```

```python
products = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 50,
    "Monitor": 12,
    "Webcam": 30}
```

```python
check_inventory(products)
```

#### Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
def process_student_results(marks):
    if not marks:
        print("No marks provided.")
        return

    total_marks = sum(marks)
    average_marks = total_marks / len(marks)

    print(f"Student marks: {marks}")
    print(f"Average marks: {average_marks:.2f}")

    if average_marks >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")
```

```python
print("--- Student 1 ---")
student1_marks = [75, 80, 65, 90, 70]
process_student_results(student1_marks)

print("\n--- Student 2 ---")
student2_marks = [40, 35, 50, 45, 30]
process_student_results(student2_marks)
```

#### Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry


```python
def cab_fare_estimator():
    base_fare = 50
    rate_per_km = 12
    peak_hour_surcharge_percentage = 0.25 # 25%

    while True:
        try:
            distance_str = input("Enter distance traveled in km: ")
            distance = float(distance_str)

            if distance <= 0:
                print("Distance must be a positive number. Please try again.\n")
                continue

            is_peak_hour_str = input("Is it peak hour? (yes/no): ").lower()

            if is_peak_hour_str not in ['yes', 'no']:
                print("Invalid input for peak hour. Please enter 'yes' or 'no'.\n")
                continue

            # Calculate base fare
            total_fare = base_fare + (distance * rate_per_km)

            # Apply peak hour surcharge if applicable
            if is_peak_hour_str == 'yes':
                total_fare *= (1 + peak_hour_surcharge_percentage)
            
            print(f"Estimated Cab Fare: ₹{total_fare:.2f}\n")
            return total_fare

        except ValueError:
            print("Invalid input. Please enter a valid number for distance.\n")
```

```python
estimated_fare = cab_fare_estimator()
print(f"Final estimated fare: ₹{estimated_fare:.2f}")
```

#### Problem 5: Employee Attendance Eligibility Checker
Problem Statement
Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement


```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        print("No attendance records provided.")
        return "Not Applicable"

    present_days = attendance_list.count('P')
    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendance records: {attendance_list}")
    print(f"Attendance percentage: {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        print("Eligibility: Eligible")
        return "Eligible"
    else:
        print("Eligibility: Not Eligible")
        return "Not Eligible"
```

#### Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function


```python
def check_password_strength(password):
    min_length = 8
    has_digit = False
    has_special_char = False
    special_chars = "@#$"

    # Rule 1: Minimum length
    if len(password) < min_length:
        print(f"Password is too short. Minimum length is {min_length} characters.")
        return "Weak"

    # Rule 2: At least one digit
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    if not has_digit:
        print("Password must contain at least one digit.")
        return "Weak"

    # Rule 3: At least one special character
    for char in password:
        if char in special_chars:
            has_special_char = True
            break
    if not has_special_char:
        print(f"Password must contain at least one special character ({special_chars}).")
        return "Weak"

    print("Password strength: Strong")
    return "Strong"
```

```python
print("--- Password 1 ---")
check_password_strength("MyPass123@")

print("\n--- Password 2 ---")
check_password_strength("Pass1$")

print("\n--- Password 3 ---")
check_password_strength("MyPassword@")

print("\n--- Password 4 ---")
check_password_strength("MyPassword123")

print("\n--- Password 5 ---")
check_password_strength("short")

print("\n--- Password 6 ---")
check_password_strength("SecureP@ss2024")
```
