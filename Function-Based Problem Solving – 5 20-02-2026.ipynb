#### Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.
Rules
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement
Real-Time Use
Social media analytics platforms
Marketing dashboards
Hint
Use function
Use loop for calculations
Expected Output Format
Total Likes: 1050
Post Status: Viral Post


```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like_count in likes_list:
        total_likes += like_count

    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    return total_likes, post_status
```

```python
likes_on_post1 = [200, 300, 550] # Example for a viral post
total_likes1, status1 = analyze_post_engagement(likes_on_post1)
print(f"Total Likes: {total_likes1}")
print(f"Post Status: {status1}")


likes_on_post2 = [100, 250, 400] # Example for a normal post
total_likes2, status2 = analyze_post_engagement(likes_on_post2)
print(f"Total Likes: {total_likes2}")
print(f"Post Status: {status2}")
```

#### Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.
Rules
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall
Real-Time Use
Climate monitoring systems
Irrigation planning tools
Hint
Use function
Use loop and math operations
Expected Output Format
Average Rainfall: 72
Rainfall Status: Adequate Rainfall


```python
def analyze_rainfall_adequacy(rainfall_measurements, required_level):
    if not rainfall_measurements:
        return 0.0, "No Rainfall Data"

    total_rainfall = sum(rainfall_measurements)
    average_rainfall = total_rainfall / len(rainfall_measurements)

    if average_rainfall >= required_level:
        rainfall_status = "Adequate Rainfall"
    else:
        rainfall_status = "Inadequate Rainfall"

    return average_rainfall, rainfall_status
```

```python
season_rainfall_1 = [60, 80, 75, 70] # Measurements for a season
required_avg_1 = 70
avg_rainfall_1, status_1 = analyze_rainfall_adequacy(season_rainfall_1, required_avg_1)
print(f"Average Rainfall: {avg_rainfall_1}")
print(f"Rainfall Status: {status_1}")

season_rainfall_2 = [50, 45, 60, 55] # Measurements for another season
required_avg_2 = 65
avg_rainfall_2, status_2 = analyze_rainfall_adequacy(season_rainfall_2, required_avg_2)
print(f"Average Rainfall: {avg_rainfall_2}")
print(f"Rainfall Status: {status_2}")

season_rainfall_3 = [70, 70, 70, 70]
required_avg_3 = 70
avg_rainfall_3, status_3 = analyze_rainfall_adequacy(season_rainfall_3, required_avg_3)
print(f"Average Rainfall: {avg_rainfall_3}")
print(f"Rainfall Status: {status_3}")
```

#### Problem 4: Social Media – Duplicate Account Detection
Problem Statement
Create a function to detect duplicate usernames in a system.
Rules
If duplicates exist → Flag accounts
Else → Accounts are unique
Real-Time Use
Platform moderation
Fraud detection systems
Hint
Use function
Use set data structure
Expected Output Format
Duplicate Accounts Found: Yes


```python
def detect_duplicate_accounts(usernames):
    
    seen_usernames = set()
    duplicate_usernames = set()
    for username in usernames:
        if username in seen_usernames:
            duplicate_usernames.add(username)
        else:
            seen_usernames.add(username)
    
    return len(duplicate_usernames) > 0, duplicate_usernames
```

```python
user_list_1 = ["user1", "user2", "user3", "user1", "user4"]
duplicates_exist_1, duplicates_1 = detect_duplicate_accounts(user_list_1)
print(f"Duplicate Accounts Found: {'Yes' if duplicates_exist_1 else 'No'}")
if duplicates_exist_1:
    print(f"Duplicated Usernames: {duplicates_1}")

user_list_2 = ["userA", "userB", "userC", "userD"]
duplicates_exist_2, duplicates_2 = detect_duplicate_accounts(user_list_2)
print(f"Duplicate Accounts Found: {'Yes' if duplicates_exist_2 else 'No'}")
if duplicates_exist_2:
    print(f"Duplicated Usernames: {duplicates_2}")

user_list_3 = ["alpha", "beta", "gamma", "alpha", "beta", "delta"]
duplicates_exist_3, duplicates_3 = detect_duplicate_accounts(user_list_3)
print(f"Duplicate Accounts Found: {'Yes' if duplicates_exist_3 else 'No'}")
if duplicates_exist_3:
    print(f"Duplicated Usernames: {duplicates_3}")
```

#### Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.
Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop
Real-Time Use
Market analytics
Export pricing systems
Hint
Use function
Use list filtering logic
Expected Output Format
Premium Crops: [2500, 3200]


```python
def filter_premium_crops(crop_prices):
    premium_crops = []
    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)
    return premium_crops
```

#### Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
Real-Time Use
Learning management systems
Internship tracking platforms
Hint
Use function
Use list and loop
Expected Output Format
Attendance Percentage: 80.0
Exam Eligibility: Eligible


```python
def check_exam_eligibility(attended_classes, total_classes):
    if total_classes == 0:
        return 0.0, "N/A (No classes)"

    attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"

    return attendance_percentage, eligibility_status
```

```python
attended_1 = 80
total_1 = 100
percentage_1, eligibility_1 = check_exam_eligibility(attended_1, total_1)
print(f"Attendance Percentage: {percentage_1}")
print(f"Exam Eligibility: {eligibility_1}")

attended_2 = 60
total_2 = 100
percentage_2, eligibility_2 = check_exam_eligibility(attended_2, total_2)
print(f"Attendance Percentage: {percentage_2}")
print(f"Exam Eligibility: {eligibility_2}")

attended_3 = 75
total_3 = 100
percentage_3, eligibility_3 = check_exam_eligibility(attended_3, total_3)
print(f"Attendance Percentage: {percentage_3}")
print(f"Exam Eligibility: {eligibility_3}")

attended_4 = 0
total_4 = 0
percentage_4, eligibility_4 = check_exam_eligibility(attended_4, total_4)
print(f"Attendance Percentage: {percentage_4}")
print(f"Exam Eligibility: {eligibility_4}")
```

#### Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
Real-Time Use
Smart meter systems
Electricity board billing software
Energy consumption analytics
Hint
Use function to calculate bill
Use conditional logic for slabs
Use return values (bill and usage status)


```python
def calculate_electricity_bill(units_consumed):
    total_bill = 0.0

    # Calculate bill based on slabs
    if units_consumed > 200:
        total_bill += (units_consumed - 200) * 7  # Units above 200 at ₹7/unit
        remaining_units = 200
    else:
        remaining_units = units_consumed

    if remaining_units > 100:
        total_bill += (remaining_units - 100) * 5  # Units between 101-200 at ₹5/unit
        remaining_units = 100

    total_bill += remaining_units * 3  # First 100 units at ₹3/unit

    # Classify usage level
    if total_bill < 500:
        usage_classification = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_classification = "Moderate Usage"
    else:
        usage_classification = "High Usage"

    return total_bill, usage_classification
```

```python
units_1 = 50
bill_1, usage_1 = calculate_electricity_bill(units_1)
print(f"Units Consumed: {units_1}")
print(f"Total Bill: ₹{bill_1}")
print(f"Usage Classification: {usage_1}")

units_2 = 150 # (100 * 3) + (50 * 5) = 300 + 250 = 550
bill_2, usage_2 = calculate_electricity_bill(units_2)
print(f"Units Consumed: {units_2}")
print(f"Total Bill: ₹{bill_2}")
print(f"Usage Classification: {usage_2}")

units_4 = 100 + 40 # 100*3 + 40*5 = 300 + 200 = 500
bill_4, usage_4 = calculate_electricity_bill(units_4)
print(f"Units Consumed: {units_4}")
print(f"Total Bill: ₹{bill_4}")
print(f"Usage Classification: {usage_4}")

units_5 = 300 # (100 * 3) + (100 * 5) + (100 * 7) = 300 + 500 + 700 = 1500
bill_5, usage_5 = calculate_electricity_bill(units_5)
print(f"Units Consumed: {units_5}")
print(f"Total Bill: ₹{bill_5}")
print(f"Usage Classification: {usage_5}")
```
